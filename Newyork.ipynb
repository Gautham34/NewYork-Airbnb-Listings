```python
top_neighbourhood = df1.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,6))
bars = plt.bar(top_neighbourhood.index, top_neighbourhood.values, color='orange')
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  
        yval + 5,                         
        f'{yval:.0f}',                    
        ha='center', va='bottom', fontsize=9
    )
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Neighbourhoods by Average Price')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price (USD)')
plt.tight_layout()
plt.show()
```

```python
min_nights_price = df1.groupby('minimum_nights')['price'].mean().sort_index().head(20)

plt.figure(figsize=(12,6))
plt.plot(min_nights_price.index, min_nights_price.values, marker='o', color='purple')
plt.title('Average Price vs Minimum Nights (Top 20)')
plt.xlabel('Minimum Nights')
plt.ylabel('Average Price (USD)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
room_counts = df1['room_type'].value_counts()

plt.figure(figsize=(10,5))
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Listings by Room Type')
plt.axis('equal')
plt.show()
```

```python
sns.barplot(data=df1, x='neighbourhood_group', y='price', hue='room_type', palette='Set1', alpha=0.6)
plt.title("Average Airbnb Price by Neighbourhood and Room Type")
plt.ylabel("Average Price")
plt.xlabel("Neighbourhood")
plt.legend(title="Room Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
sns.scatterplot(data=df1, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title("NYC Airbnb: Price vs Number of Reviews by Neighbourhood Group")
plt.ylabel("Price in USD")
plt.xlabel("No of Reviews")
plt.legend(title="Neighbourhood Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data=df1, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type', palette='Set2',
    plot_kws={'alpha':0.5, 's':25})
plt.suptitle("Room Type Analysis: Key Numeric Features", fontsize=14, fontweight="bold", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df1, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of Airbnb")
plt.ylabel("Latitude")
plt.xlabel("Longitude")
plt.legend(title='Room Type')
plt.show()
```

```python
corr = df1[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr
plt.figure(figsize=(8,6))
sns.heatmap(data=corr, annot=True, cmap="BrBG", center=0, fmt=".2f")
plt.title("Correlation Heatmap of Airbnb Dataset", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```


## KEY INSIGHTS

**Price Distribution:**
Most Airbnb listings are priced below 500 Dollars, with a few high-priced properties forming a long tail. Listings above 1500 Dolllars were excluded to focus on the majority price range and remove extreme outliers. 

**Availability:**
The majority of listings have limited annual availability, indicating that many properties are either frequently booked or only listed seasonally.

**Top Neighbourhoods:**
The most expensive neighbourhoods on average are concentrated in central Manhattan. Neighbourhoods like Tribeca, Midtown, and Theatre District command significantly higher average prices compared to outer boroughs.

**Minimum Nights vs Price:**
Listings with higher minimum night requirements tend to have **higher average prices**, indicating that premium or luxury stays often enforce longer booking durations.

**Room type Distribution:**
Entire homes/apartments make up the largest share of listings, followed by private rooms. Shared rooms form a very small portion of the market, highlighting that most hosts offer independent spaces.

**Neighbourhood Group and Room Types:**
Entire homes/apartments in Manhattan have the highest average prices. Private and shared rooms are more affordable and spread across other boroughs like Brooklyn and Queens.

**Price vs Number of Reviews:**
Listings with higher numbers of reviews are generally lower-priced, suggesting that affordable stays attract more guests and accumulate more reviews over time.

**Multivariate Analysis(Pairplot):**
Different room types show distinct patterns: entire homes typically have higher prices and minimum nights, while private rooms often have higher review counts and availability.

**Geographical Distribution:**
High-priced listings are clustered in central Manhattan, with room types showing clear spatial patterns — entire homes dominate in premium locations, while private rooms are more scattered.

**Correlation Heatmap:**
Price shows weak correlations with most numeric variables like number of reviews and availability, indicating that location and room type are stronger determinants of price than purely numeric metrics. Latitude and longitude have indirect spatial effects influencing price.
